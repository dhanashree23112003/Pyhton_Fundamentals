Q2 Take an input from the user and print the factorial of that number. Note: A factorail is the function that multiplies a number by every number below it.

```python
num = int(input("Enter a number"))
var = 1

for i in range(1, num+1):
  if num < 0:
    print("Invalid Input")
  elif num == 0 or num == 1:
    print("Factorial is 1")
  else:
    var = var * i
print(var)
```

```python
num = int(input("Enter a number"))
total = 1

while num > 0:
  total = total * num
  num = num - 1
print(total)
```

```python
num = int(input("Enter a number"))
total = 1

for i in range(num,0,-1):
  total = total * i
print(total)
```

Q4: Given a string, write a python function if its palindrome or not. A string is said to be Palindrome if the reverse of the string is as string. For example, radar in reverse is radar, not radix, is not palindrome

```python
var = input("Enter anything ")

if var == var[::-1]:
    print("Palindrome")
else:
    print("Not Palindrome")
```

Q5: Make a two player Rock - Paper - Scissors game. HInt: Ask for players play(using input), compare them, print a message of congraturaltion

```python
series = int(input("Enter the number of series"))

player1_arr = 0
player2_arr = 0

for i in range(series):
   Player1 = input("Enter your choice (R, P, S)").upper()
   Player2 = input("Enter your choice (R, P, S)").upper()

   if Player1 == Player2:
    print("Draw")

   elif (Player1 == "R" and Player2 == "S") or (Player1 == "S" and Player2 == "R") or (Player1 == "P" and Player2 == "R"):
    print("Player1 wins")
    player1_arr += 1

   else:
    print("Player2 wins")
    player2_arr += 1

if player1_arr > player2_arr:
  print("Player1 wins")
elif player2_arr > player1_arr:
  print("Player2 wins")
else:
  print("Draw")
```
